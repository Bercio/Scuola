##Pendenza di una retta

La pendenza di una curva é determinata dal coefficiente angolare della retta
tangente in quel punto, cambia quindi di punto in punto.


```python
%pylab inline
x = pylab.linspace(0, 1, 1000)
y = x**2
pylab.figure(figsize=(10,10))
pylab.plot(x,y)
A = 0.4
B = 0.7
scatter(A,A**2, 20, color='red')
scatter(B,B**2, 20, color='red')
pylab.plot([B,B],[A**2,B**2], color='red', linestyle="--")
pylab.plot([B,A],[A**2,A**2], color='red', linestyle='--')
pylab.plot([A,B],[A**2,B**2], color='red')
pylab.annotate('A', xy=(A, A**2), xycoords='data',xytext=(+5, +15), textcoords='offset points')
pylab.annotate('B', xy=(B, B**2), xycoords='data',xytext=(+0, +15), textcoords='offset points')
pylab.ylim(0,1)
pylab.xlim(0,1)
pylab.show()
```
